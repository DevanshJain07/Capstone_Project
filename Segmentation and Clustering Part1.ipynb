```python
url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

```python
soup = BeautifulSoup(url, 'lxml')
```

```python
table_post = soup.find('table')
td = table_post.find_all('td')

postcode = []
borough = []
neighbourhood = []

for i in range(0, len(td), 3):
    postcode.append(td[i].text.strip())
    borough.append(td[i+1].text.strip())
    neighbourhood.append(td[i+2].text.strip())
        
toronto = pd.DataFrame(data=[postcode, borough, neighbourhood]).transpose()
toronto.columns = ['Postcode', 'Borough', 'Neighbourhood']
toronto.head()
```

```python
toronto['Borough'].replace('Not assigned', np.nan, inplace=True)
toronto.dropna(subset=['Borough'], inplace=True)

toronto.head()
```

```python
toronto = toronto.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(', '.join).reset_index()
toronto.columns = ['Postcode', 'Borough', 'Neighbourhood']
toronto.head(20)
```
